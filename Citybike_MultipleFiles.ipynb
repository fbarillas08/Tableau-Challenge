```python
## Convert trip duration to minutes
JC2019['tdur_minutes']=JC2019['tripduration']/60
JC2020['tdur_minutes']=JC2020['tripduration']/60
JC2021['tdur_minutes']=JC2021['tripduration']/60
```

```python
## Convert trip duration to hours
JC2019['tdur_hours']=JC2019['tripduration']/3600
JC2020['tdur_hours']=JC2020['tripduration']/3600
JC2021['tdur_hours']=JC2021['tripduration']/3600
```

```python
## Convert trip duration to days
JC2019['tdur_days']=JC2019['tdur_hours']/24
JC2020['tdur_days']=JC2020['tdur_hours']/24
JC2021['tdur_days']=JC2021['tdur_hours']/24
```

```python
## Convert start and end times to datetime for parseling
JC2019['starttime']=pd.to_datetime(JC2019.starttime)
JC2019['stoptime']=pd.to_datetime(JC2019.stoptime)

JC2020['starttime']=pd.to_datetime(JC2020.starttime)
JC2020['stoptime']=pd.to_datetime(JC2020.stoptime)

JC2021['starttime']=pd.to_datetime(JC2021.starttime)
JC2021['stoptime']=pd.to_datetime(JC2021.stoptime)
```

```python
JC2019['start_time']=JC2019['starttime'].dt.time
JC2019['stop_time']=JC2019['stoptime'].dt.time

JC2019['start_day']=JC2019['starttime'].dt.date
JC2019['stop_day']=JC2019['stoptime'].dt.date

JC2020['start_time']=JC2020['starttime'].dt.time
JC2020['stop_time']=JC2020['stoptime'].dt.time

JC2020['start_day']=JC2020['starttime'].dt.date
JC2020['stop_day']=JC2020['stoptime'].dt.date

JC2021['start_time']=JC2021['starttime'].dt.time
JC2021['stop_time']=JC2021['stoptime'].dt.time

JC2021['start_day']=JC2021['starttime'].dt.date
JC2021['stop_day']=JC2021['stoptime'].dt.date
```

```python
## Find the day of the week bike was rented
JC2019['start_day']=pd.to_datetime(JC2019.start_day)
JC2019['DOW_rented']=JC2019['start_day'].dt.day_name()

JC2020['start_day']=pd.to_datetime(JC2020.start_day)
JC2020['DOW_rented']=JC2020['start_day'].dt.day_name()

JC2021['start_day']=pd.to_datetime(JC2021.start_day)
JC2021['DOW_rented']=JC2021['start_day'].dt.day_name()
```

```python
## Convert gender codes to categories
JC2019.loc[JC2019.gender==1, "gender"] = "Male"
JC2019.loc[JC2019.gender==2, "gender"] = "Female"
JC2019.loc[JC2019.gender==0, "gender"] = "Unknown"

JC2020.loc[JC2020.gender==1, "gender"] = "Male"
JC2020.loc[JC2020.gender==2, "gender"] = "Female"
JC2020.loc[JC2020.gender==0, "gender"] = "Unknown"

JC2021.loc[JC2021.gender==1, "gender"] = "Male"
JC2021.loc[JC2021.gender==2, "gender"] = "Female"
JC2021.loc[JC2021.gender==0, "gender"] = "Unknown"
```
